## 投手數據整理

```python
excel_file_path = 'C:/Users/ASUS/Documents/實習/Rebas/CPBL_2023.xlsx'
sheet_name3 = 'AWAY-BOX-投球'  # 請替換成你要載入的工作表名稱
sheet_name4 = 'HOME-BOX-投球'
away1 = pd.read_excel(excel_file_path, sheet_name=sheet_name3)
home1 = pd.read_excel(excel_file_path, sheet_name=sheet_name4)
```

```python
away1.rename(columns={'客場': '球隊'}, inplace=True)
home1.rename(columns={'主場': '球隊'}, inplace=True)
```

```python
merged_stat = pd.concat([away1, home1], ignore_index=True)

#merged_stat = merged_stat.drop(["比賽編號","比賽日期","客場","主場","背號"], axis = 1)
merged_stat
```

```python
agg_dict = {
    "出局數": "sum",
    "面對打者": "sum",
    "安打": "sum",
    "全壘打": "sum",
    "保送": "sum",
    "故意四壞": "sum",
    "觸身": "sum",
    "三振": "sum",
    "暴投": "sum",
    "投手犯規": "sum",
    "失分": "sum",
    "責失": "sum"
}
pitcher_total = merged_stat.groupby(["名稱",'球隊']).agg(agg_dict).reset_index()
pitcher_total
```

```python
#2023 ERA:3.705 FIP+:3.791 
#2022 ERA:3.347 FIP+:3.431
pitcher_total["IP"] = pitcher_total["出局數"] / 3
pitcher_total["ERA"] = (pitcher_total["責失"] / pitcher_total["IP"]) * 9
pitcher_total["FIP"] = 0.33 * (3 *pitcher_total["全壘打"] + 3 * pitcher_total["保送"] + 2 * pitcher_total["觸身"] + 3 * pitcher_total["暴投"] - 2 * pitcher_total["三振"]) / pitcher_total["IP"] + 3.2
pitcher_total["K%"] = (pitcher_total["三振"] / pitcher_total["面對打者"]) * 100
pitcher_total["BB%"] = (pitcher_total["保送"] / pitcher_total["面對打者"]) * 100
pitcher_total["ERA+"] = (3.705 / pitcher_total["ERA"]) * 100
pitcher_total["FIP+"] = (3.791 / pitcher_total["FIP"]) * 100

# 移除指定的欄位
pitcher_total.drop(columns=["出局數","暴投","投手犯規","故意四壞","觸身"], inplace=True)
pitcher_total = pitcher_total.rename(columns={'名稱': '投手'})

# 列出計算結果
pitcher_total
```

## 投手成績和tRA合併

```python
excel_file_path2 = 'C:/Users/ASUS/Documents/實習/WAR/WAR2.0/2023tRA.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name2 = 'Sheet1'
item = pd.read_excel(excel_file_path2, sheet_name=sheet_name2)
```

```python
item
```

```python
all_D = pd.merge(pitcher_total, item, on=['投手','球隊'], how='inner')
all_D
```

```python
pitcher = all_D[['投手', '球隊','IP','ERA','FIP', 'HBP', 'HR', 'SO', 'uBB', '內野飛球', '外野飛球', '平飛球', '滾地球','tRA']]
pitcher
```

```python
park_factors = {
    "味全龍": 0.13360794745482,
    "統一7-ELEVEn獅": 0.133922753699232,
    "富邦悍將": 0.134623205953924,
    "樂天桃猿": 0.134173092008184,
    "中信兄弟":0.13580124761886
}

# 使用球场名称从字典中获取 "Park Factor" 值
pitcher["PF(tRA)"] = pitcher["球隊"].map(park_factors)

# 打印结果
pitcher
```

```python
#2023 聯盟tRA=3.436941872
#2022 聯盟tRA=3.347000
#RAR＝（1.356×聯盟平均tRA－PF補正後tRA）÷9×（局數）
lg_tr=3.700000
pitcher["RAR"]=((1.365*lg_tr-pitcher["tRA"]*pitcher["PF(tRA)"])/9*pitcher["IP"])

#2023 聯盟RPW(FG)=9.35166     
#2023 聯盟RPW(1.02)=6.8593172473
#2022 聯盟RPW(FG)=9.102359373
#2022 聯盟RPW(1.02)=6.723292499
pitcher["WAR(FG)"]=pitcher["RAR"]/9.35166
pitcher["WAR(1.02)"]=pitcher["RAR"]/6.8593172473
pitcher
```
